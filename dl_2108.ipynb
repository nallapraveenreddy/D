```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision.datasets import CIFAR10
from torch.utils.data import DataLoader
from torchvision import transforms
import torch.optim as optim
```

```python
transforms = transforms.Compose([transforms.ToTensor(),transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))])
```

```python
train_data = CIFAR10(root = './data',download = True,train =True, transform = transforms)
test_data = CIFAR10(root = './data',download = True,train =False, transform = transforms)
```

```python
trainloader = DataLoader(train_data, batch_size = 32,shuffle = True)
testloader = DataLoader(train_data, batch_size = 32,shuffle = False)
```

lenet-5.svg

```python
class Cifar_CNN(nn.Module):
  def __init__(self):
    super(Cifar_CNN,self).__init__()
    self.conv1 = nn.Conv2d(3,6,5)
    self.conv2 = nn.Conv2d(6,16,5)
    self.fc1 = nn.Linear(5*5*16,120)
    self.fc2 = nn.Linear(120,84)

  def forward(self,x):
    x = self.conv1(x)
    x = F.relu(x)
    x = F.max_pool2d(x,2)

    x = self.conv2(x)
    x = F.relu(x)
    x = F.max_pool2d(x,2)

    x=x.view(-1,5*5*16)

    x = self.fc1(x)
    x = F.relu(x)

    x = self.fc2(x)
    return x
```

```python
obj = Cifar_CNN()
criterian = nn.CrossEntropyLoss()
optimizer = optim.Adam(obj.parameters(),lr = 0.001)
```

```python
for epoch in range(5):
  loss_epoch = 0
  total = 0
  correct = 0
  for images,label in trainloader:
    predictions = obj(images)
    loss = criterian(predictions,label)
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()
    _, predicted = torch.max(predictions.data,1)
    total += label.size(0)
    correct += (predicted==label).sum().item()
    loss_epoch += loss.item()
  print(f"Accuracy:{(correct/total)*100}")
```

```python
obj.eval()
correct = 0
total = 0
with torch.no_grad():
  for images,labels in testloader:
    predictions = obj.forward(images)
    _,predicted = torch.max(predictions.data,1)
    correct += (predicted==labels).sum().item()
    total += labels.size(0)
  print(f"Accuracy:{(correct/total)*100}")
```
